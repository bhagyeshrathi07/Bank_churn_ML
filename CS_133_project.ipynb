```python
# Importing the data and cleaning
import pandas as pd

bank_churn = pd.read_csv("/content/Customer-Churn-Records.csv", index_col="RowNumber", na_values="-", decimal=".")
bank_churn.head(20)
```

```python
# Drop unnecessary column
bank_churn.drop(columns=['Surname'], inplace=True)

# Rename columns for clarity
bank_churn = bank_churn.rename(columns={"Geography": "Country", "Card Type": "Card_Type", "Satisfaction Score": "Satisfaction_Score", "Point Earned": "Points_Earned"})

bank_churn
```
